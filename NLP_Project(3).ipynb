```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

checkpoint = "sshleifer/distilbart-cnn-12-6"

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
```

```python
from langchain.text_splitter import TokenTextSplitter
```

```python
!pip install langchain
```

```python
def chunk_text(text):

    splitter = TokenTextSplitter(
        encoding_name="gpt2",
        chunk_size=600,
        chunk_overlap=20
    )
    output = splitter.create_documents([text])
    return output
```

```python
import os
directory = r'/content/IN-Abs/train-data/judgement'
files = ['1.txt', '10.txt', '100.txt','1000.txt', '1001.txt','1002.txt','1003.txt', '1004.txt', '1005.txt', '1006.txt']
summaries_bert=[]
for i in files:
    file_path = os.path.join(directory, i)
    print(file_path)
    with open(file_path, 'r') as file:
        contents = file.read()
        chunks=chunk_text(contents)
        print(len(chunks))
        inputs = [tokenizer(str(chunk), return_tensors="pt") for chunk in chunks]
        for input in inputs:
          output = model.generate(**input)
          a=tokenizer.decode(output[0], skip_special_tokens=True)
          print(a)
          summaries_bert.append(a)
        print("....................")
```

# **T5**

```python
import os
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM  # Importing tokenizer and model

# Define the directory and list of files
directory = r'/content/IN-Abs/train-data/judgement'
files = ['1.txt', '10.txt', '100.txt', '1000.txt', '1001.txt','1002.txt', '1003.txt', '1004.txt', '1005.txt', '1006.txt']

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
summaries_t5=[]
# Iterate over each file
for i in files:
    file_path = os.path.join(directory, i)
    print(file_path)

    # Read the contents of the file
    with open(file_path, 'r') as file:
        contents = file.read()

    # Tokenize the text and generate summary
    input_tokenized = tokenizer.encode(contents, return_tensors='pt', padding='longest', pad_to_max_length=True, max_length=6144, truncation=True)
    summary_ids = model.generate(input_tokenized,
                                  num_beams=4,
                                  no_repeat_ngram_size=3,
                                  length_penalty=2,
                                  min_length=350,
                                  max_length=500)
    summary = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids][0]
    summaries_t5.append(summary)
    # Print the generated summary
    print("Summary for " + i)
    print(summary)
    print("................................")
```

## **BLUE Score**

```python
import nltk

def calculate_bleu(reference, candidate):
    reference = [reference.split()]
    candidate = candidate.split()
    return nltk.translate.bleu_score.sentence_bleu(reference, candidate)
for i in range(len(files)):
    print(files[i])
    bleu_score = calculate_bleu(summaries_t5[i],summaries_text[i])
    print("BLEU Score:", bleu_score)
    print("....................................................")
```

## **ROGUE Score**

```python
import nltk
import rouge
from rouge import Rouge

def calculate_rouge(reference, candidate):
    rouge = Rouge()
    scores = rouge.get_scores(candidate, reference)
    return scores[0]['rouge-l']['f']
for i in range(len(files)):
    print(files[i])
    rouge_score = calculate_rouge(summaries_t5[i],summaries_text[i])
    print("Rouge Score:", rouge_score)
    print("....................................................")
```
